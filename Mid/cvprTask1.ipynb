```python
import os
import cv2

# Define the root directory where the CIFAR-10 images are located
cifar10_root = r'C:\Users\User\Desktop\CVPR\Mid\train'

# List the subdirectories (class folders)
class_folders = os.listdir(cifar10_root)

# Initialize empty lists to store grayscale images and labels
gray_image_data = []
labels = []

# Loop through each class folder
for class_folder in class_folders:
    class_folder_path = os.path.join(cifar10_root, class_folder)
    
    # Check if it's a directory
    if os.path.isdir(class_folder_path):
        # Loop through images in the class folder
        for image_file in os.listdir(class_folder_path):
            image_path = os.path.join(class_folder_path, image_file)
            
            # Read the image using cv2
            image = cv2.imread(image_path)
            
            # Check if the image was successfully loaded
            if image is not None:
                # Convert the image to grayscale
                gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
                
                # Append the grayscale image data to the list
                gray_image_data.append(gray_image)
                
                # Append the class label (class folder name) to the labels list
                labels.append(class_folder)
            else:
                print(f"Failed to load image: {image_path}")
            
```
